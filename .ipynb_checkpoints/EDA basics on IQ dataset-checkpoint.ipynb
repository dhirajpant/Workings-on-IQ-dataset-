```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing dataset
df=pd.read_csv("International_IQ.csv")
```

```python
#shows first 10 rows of dataset
df.head(10)
```

```python
#total no of rows and columns
df.shape
```

```python
#checking null or missing values
df.isnull().sum()
```

```python
#delete null values
df=df.dropna()
```

```python
df.info()
```

```python
df.describe()
```

```python
#sort out top 10 countries with higest iq
most_iq_countries=df.sort_values(by='Average IQ',ascending=False).head(10)
most_iq_countries
```

```python
#sort out top 10 countries with most participants
most_participants_countries=df.sort_values(by='Participants',ascending=False).head(10)
most_participants_countries
```

```python
#plotting a bar graph with top 10 countries by average iq
plt.figure(figsize=(18, 10))
plt.bar(df["Country"].head(10), df["Average IQ"].head(10))
plt.xlabel("Average IQ")
plt.title("Top 10 Countries by Average IQ")

#Display the plot
plt.show()
```

```python
#created a new dataframe by sorting the no of participants in ascending order
df1=df.sort_values(by='Participants',ascending=False).head(10)
df1
```

```python
#plotting a bar graph with top 10 countries by most participants
plt.figure(figsize=(18, 10))
plt.bar(df1["Country"].head(10), df1["Participants"].head(10), color='skyblue')
plt.xlabel("Participants")
plt.title("Top 10 Countries by Average IQ")

# Display the plot
plt.show()
```
